```python
rows = []
i = 0
for row in df[['ID','identifier']].iterrows():
    #i += 1
    #if i > 5:
    #    break
    
    if i % 100 == 0:
        print('processed', i)
    i += 1
    rid = row[1][0]
    ridentifier = row[1][1]
    
    #print(rid, ridentifier)
    
    item = get_item(ridentifier)
    download(ridentifier, glob_pattern='*files.xml', verbose=True)

    file = ridentifier + '/' + ridentifier + '_files.xml'

    with open(file) as f:

        jsn = xmltodict.parse(f.read())
        for f in jsn['files']['file']:  
            if f['@name'].split('.')[-1] == 'mp4':  
                dl = f['@name']
                s = int(f['size'])
                url = 'https://archive.org/download/' + file.split('/')[1] + '/' + dl
                identifier = file.split('/')[0]
                row = []
                row.append(rid)
                row.append(identifier)
                row.append(dl)
                row.append(s)
                row.append(url)
        rows.append(row)
        shutil.rmtree(file.split('/')[0])

print(rows)
```

```python
df_urls = pd.DataFrame(rows)
df_urls.columns = ['id', 'identifier', 'file_name', 'file_size', 'download_url']
```
