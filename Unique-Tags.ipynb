```python
import pandas as pd
from pandasql import sqldf
from collections import defaultdict
pysqldf = lambda q: sqldf(q, globals())
pd.set_option('display.max_rows', 500)
```

```python
df_human = pd.read_csv("data/classifications.csv")
```

```python
# all unique tags
unique_tags = set()
for row in df_human.iterrows():
    #print(row[1]['subject'])
    tags = row[1]['subject'].split(';')
    for t in tags:
        t = t.strip()
        if t != '' and 'href' not in t:
            unique_tags.add(t.strip())
```

```python
df_ml = pd.read_csv("data/transcript_accuracy.csv")
```

```python
df_dataset = pysqldf(f"""
SELECT dh.*
FROM 
    df_human dh
JOIN
    df_ml dml
ON dh.ID = dml.id
    WHERE description is not NULL
    AND description != 'error code 224003'
""")
```

```python
# all unique tags
limited_tags = defaultdict(int)
for row in df_dataset.iterrows():
    #print(row[1]['subject'])
    tags = row[1]['subject'].split(';')
    for t in tags:
        t = t.strip()
        if t != '' and 'href' not in t:
            limited_tags[t] += 1
```

```python
df = pd.DataFrame(limited_tags.items(), columns=['Tag', 'Count'])
```
